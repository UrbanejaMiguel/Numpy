### Adding, removing and sorting elements

```python
import numpy as np
```

```python
first_arr=np.arange(1,11)
first_arr
```

```python
new_first_arr=np.append(first_arr,11)
new_first_arr
```

```python
second_arr=np.arange(12,22)
second_arr
```

```python
new_second_arr=np.insert(second_arr,(0,10),[-100,-100])
new_second_arr
```

```python
third_arr=np.arange(1,6)
third_arr
```

```python
del_arr=np.delete(third_arr,2)
del_arr
```

```python
fourth_arr=np.random.randint(1,100,100)
fourth_arr
```

```python
sort_arr=np.sort(fourth_arr)
sort_arr

#Ordenar de mayor a menor se hace asi: 
sort_arr_2=np.sort(fourth_arr)[::-1]
sort_arr_2
```

```python
fiveth_arr=np.random.randint(1,50,(100,100))
fiveth_arr #Te pone el shape porque no muestra toda la matriz, entonces te indica el tama√±o que tiene
```

```python
sort_arr3=np.sort(fiveth_arr)
sort_arr3
```

```python
sort_arr4=np.sort(fiveth_arr,0)
sort_arr4
```

```python
colors=np.array(['orange','green','yellow','white','black','pink','blue'])
colors
```

```python
sort_colors=np.sort(colors)
print(sort_colors)
```
