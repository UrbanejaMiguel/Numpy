### Reshaping arrays

```python
import numpy as np
```

```python
first_arr=np.arange(1,13)
print(first_arr)
```

```python
reshape_arr=first_arr.reshape(3,4)
print(reshape_arr)
```

```python
reshape_arr2=first_arr.reshape(3,2,2 )
print(reshape_arr2)
print("\nDimensions of reshape_arr2 is",reshape_arr2.ndim)
```

```python
second_arr=np.array([[1,2],[3,4],[5,6]])
second_arr
```

```python
reshape_arr3=second_arr.reshape(-1)
reshape_arr3
```

Hay otras maneras de aplanar un array. Se puede usar reval o flatten. Reval es view con lo cual cambia el original y es mas eficiente. Sin embargo, flatten crea una copia aplanada.

```python
arr=np.arange(1,13).reshape(3,4)
print(arr)
arr_ravel=np.ravel(arr)
print("arr_ravel",arr_ravel, arr_ravel.shape)
print("arr",arr, arr.shape)
arr_flattem = arr.flatten()
print("arr_flattem",arr_flattem, arr_flattem.shape)
print("arr",arr, arr.shape)
```
