### Joining and splitting ararys

```python
import numpy as np
```

```python
arr1=np.array([1,2,3,4,5])
arr2=np.array([6,7,8,9,10])
print(arr1)
print(arr2)
arr= np.concatenate((arr1,arr2))
print(arr)
```

```python
arr3=np.array([[1,2,3],[4,5,6]])
print(arr3)
arr4=np.array([[7,8,9],[10,11,12]])
print(arr4)
arr5=np.concatenate((arr3,arr4),axis=0)
print(arr5)
arr6=np.concatenate((arr3,arr4),axis=1)
print(arr6)
```

```python
st_arr=np.stack((arr1,arr2))
print(st_arr)
```

```python
st_arr2=np.stack((arr1,arr2),axis=1)
print(st_arr2)
```

```python
vst_arr=np.vstack((arr1,arr2))
print(vst_arr)
```

```python
hst_arr=np.hstack((arr1,arr2))
print(hst_arr)
```

```python
fifth_arr = np.arange(1,13)
print(fifth_arr)
```

```python
sp_arr=np.split(fifth_arr,2)
print(sp_arr)
```

```python
print(sp_arr[1])
```

```python
sp_arr2=np.array_split(fifth_arr,5)
sp_arr2

#ARRAY_SPLIT() TE SPLITEA LA MATRIZ EN 5 AUNQUE NO DE EXACTO
```

```python
hs_arr=np.hsplit(fifth_arr,2)
print(hs_arr)
```

```python
vst_arr=np.vsplit(fifth_arr.reshape(3,4),3) 
print(vst_arr)
```
