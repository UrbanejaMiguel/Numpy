### Indexing and slicing

```python
import numpy as np
```

```python
arr=np.arange(12)
arr
```

```python
arr=arr.reshape(3,4)
arr
```

```python
print(arr[1,1])
```

```python
print(arr[0])
```

```python
arr3d=np.reshape(np.arange(3*4*5),(3,4,5))
arr3d
```

```python
print(arr3d[0,2,3])
```

```python
print(arr3d[-1,-1,-1])
```

```python
print(arr3d[:2,:2,:2])
```

```python
print(arr3d[::,::-1,::-1])
```
