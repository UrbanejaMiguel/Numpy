### Transpose like operations

```python
import numpy as np
```

```python
arr1_2d=np.arange(12).reshape(3,4)
arr1_2d
```

```python
np.transpose(arr1_2d)
```

```python
arr2_3d=np.arange(8).reshape(2,2,2)
arr2_3d
```

```python
print(np.transpose(arr2_3d))
print("\n")
print(np.transpose(arr2_3d,(0,1,2)))
print("\n")
print(np.transpose(arr2_3d,(0,2,1)))
print("\n")
print(np.transpose(arr2_3d,(1,0,2)))
print("\n")
print(np.transpose(arr2_3d,(1,2,0)))
```

```python
np.moveaxis(arr2_3d,0,-1)
```

```python
np.swapaxes(arr2_3d,0,1)
```
