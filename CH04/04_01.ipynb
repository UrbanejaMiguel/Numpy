### Arithmetic operations

```python
import numpy as np
```

```python
a= np.arange(1,11)
b=np.arange(21,31)
print(a)
print(b)
```

```python
a+b
```

```python
b-a
```

```python
a*b
```

```python
a/b
```

```python
a**b
```

```python
print(np.add(a,b))
print(np.subtract(b,a))
print(np.multiply(a,b))
print(np.divide(a,b))
print(np.power(a,b))
```
