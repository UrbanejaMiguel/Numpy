### How to get unique items and counts

```python
import numpy as np
```

```python
arr1=np.random.randint(1,10,20)
arr1
```

```python
np.unique(arr1)
```

```python
arr2=np.array([[1,2,1,4,5],[1,7,1,9,10],[1,2,1,4,5],[1,4,1,6,7]])
arr2
```

```python
np.unique(arr2)
```

```python
np.unique(arr2,return_counts=True)
```

```python
np.unique(arr2,axis=0) #Devuelve las filas únicas
```

```python
np.unique(arr2,axis=1) #Devuelve las columnas únicas
```

```python
np.unique(arr2,axis=0,return_index=True)
```
