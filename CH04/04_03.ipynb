### Aggregate functions

```python
import numpy as np
```

```python
arr1=np.arange(10,110,10)
arr2=np.arange(10,100,10).reshape(3,3)
arr3=np.arange(10,110,10).reshape(2,5)
print(arr1)
print(arr2)
print(arr3)
```

```python
arr1.sum()
```

```python
print(arr2.sum(axis=0)) #sum of each column
print(arr2.sum(axis=1)) #sum of each row
```

```python
arr1.prod()
```

```python
print(arr3.prod(axis=0)) #product of each column
print(arr3.prod(axis=1)) #product of each row
```

```python
np.average(arr1)
```

```python
print(arr2.min(),arr2.max())
```

```python
print(arr1.mean(),np.std(arr1))
```
