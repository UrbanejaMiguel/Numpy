### Intrisic NumPy array creation

```python
import numpy as np
```

```python
integers = np.arange(10)
integers
```

```python
integers_2= np.arange(0,10)
integers_2
```

```python
integers_3= np.arange(0,10,2)
integers_3
```

```python
integers_4= np.linspace(0,5)
integers_4 #Hace una secuencia de 50 elementos
```

```python
integers_5= np.linspace(0,100,5)
integers_5
```

```python
integers_6= np.linspace(0,100,5,endpoint=False)
integers_7= np.linspace(0,100,5,endpoint=True)
print(integers_6,integers_7)
```

```python
random_nums = np.random.rand(10,4)
random_nums
```

```python
random_ints = np.random.randint(0,100,10)
random_ints
```
