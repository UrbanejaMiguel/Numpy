### Array types and conversions between types

```python
import numpy as np
```

```python
integers = np.array([10,20,30,40,50])
print(integers)
```

```python
integers[0]
```

```python
integers[0] = 100
integers
```

```python
integers[0]=21.8
integers
#Trunca a int porque la tupla es de enteros
```

```python
integers.dtype
```

```python
integers = integers.astype(np.float128)
print(f"Aunque ya se le indicara que era 21.8 al ser int, no se comvierte cuando la ponemos como float{integers}")

integers[0] = 21.8
print(integers)
```

```python
smaller_integers = np.array(integers, dtype=np.int8)
print(integers.nbytes,smaller_integers.nbytes)
```

```python
floats=np.array([1.2,2.3,3.4,5.1,8.3])
print(floats)
print(floats.dtype)
#Numpy automaticamente lo convierte a float64
```
